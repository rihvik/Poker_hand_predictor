Importing Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Importing Data

```python
train_df = pd.read_csv("/content/drive/MyDrive/poker-hand-training-true.data")
train_df
```

Pre-Processing Data

```python
test_df = pd.read_csv("/content/drive/MyDrive/poker-hand-testing.data")
test_df
```

```python
attributes=['Suit of card #1','Rank of card #1','Suit of card #2','Rank of card #2','Suit of card #3','Rank of card #3','Suit of card #4','Rank of card #4','Suit of card #5','Rank of card 5','Poker Hand']
train_df.columns = attributes
test_df.columns = attributes
```

```python
Y_train = train_df['Poker Hand']
Y_test = test_df['Poker Hand']
X_train = train_df.drop('Poker Hand',axis=1)
X_test = test_df.drop('Poker Hand',axis=1)
```

```python
X_train
```

```python
Y_train
```

Training the model using Logistic Regression

```python
model = LogisticRegression(max_iter=10000)
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(accuracy)
```
